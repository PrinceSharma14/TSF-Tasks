# Heatmap : To represent correlation between entities

```python
import seaborn as sns
corr = df.corr()
sns.heatmap(corr,annot=True)
```

```python
import numpy as np
s1 = pd.Series(np.array(df['Country']))
print(s1.describe())
```

```python
s2 = pd.Series(np.array(df['City']))
print(s2.describe())
```

```python
s3 = pd.Series(np.array(df['Category']))
print(s3.describe())
```

```python
import matplotlib.pyplot as plt
sns.countplot(df['Segment'])
plt.show()
```

# Histogram 

```python
df.hist(figsize=(10,10),bins=50)
plt.show()
```

```python
df1 = df[['City','Category', 'State','Sales','Profit']]
# pandas pivot with multiple variables
heatmap1_data = pd.pivot_table(df1,values='Sales', index=["City"], columns='Category')
```

# Heatmap of sales in every city by category

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap1_data, cmap="BuGn")
```

# Heatmap of sales in every State by category

```python
heatmap2_data = pd.pivot_table(df1,values='Sales', index=["State"], columns='Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap2_data, cmap="BuGn")
```
